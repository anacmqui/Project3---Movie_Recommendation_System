**1st Table - Get the movies for kids for the top studios**

```python
## Conditions to filter the movies
cond2 = movies_1['tomatometer_status']!= 'Rotten' #Movies that have good rating
cond3 = movies_1['genres'].str.contains('Kids')
mov_kids = movies_1[cond2 & cond3]
mov_kids['production_company'].value_counts().head(20)
```

```python
## Create function to group together the same studios which are written in different ways

def producer(x):
    if 'warner' in str(x).lower() :
        return 'Warner Bros.'
    elif 'disney' in str(x).lower() :
        return 'Disney'
    elif 'mgm' in str(x).lower():
        return 'MGM'
    elif 'fox' in str(x).lower():
        return '20th Century Fox'
    elif 'paramount' in str(x).lower():
        return 'Paramount'
    elif 'buena vista' in str(x).lower():
        return 'Buena Vista'
    elif 'universal' in str(x).lower():
        return 'Universal'
    elif 'sony' in str(x).lower():
        return 'Sony'
    elif 'miramax' in str(x).lower():
        return 'Miramax'
    elif 'dreamworks' in str(x).lower():
        return 'Dreamworks'
    else:
        return str(x)
    
mov_kids['studio'] = mov_kids['production_company'].apply(producer)
mov_kids['studio'].value_counts()
moviesperstudio = pd.DataFrame(mov_kids['studio'].value_counts()).reset_index()
moviesperstudio.head(20)
```

```python
dfstudios = mov_kids.merge(moviesperstudio, how='inner', left_on='studio', right_on='index')
dfstudios
```

```python
dfstudios['tomatometer_count'].describe()
```

```python
dfstudios['studio_y'].describe()
```

```python
cond_a = dfstudios['studio_y']>6 #get studios which produced more than 6 movies
cond_b = dfstudios['tomatometer_count']>35 #median values
top_studios = pd.DataFrame(dfstudios[cond_a & cond_b][['rotten_tomatoes_link','movie_title','tomatometer_rating','studio_x','original_release_date']])

top_studios['original_release_date'].fillna(2001, inplace = True) #there was 1 movies without year so we checked online to get the year and filled it our table

#convert 'original_release_date' into a new column as an integer
top_studios['year_release'] = top_studios['original_release_date'].astype(str)
top_studios['year_release'] = top_studios['year_release'].apply(lambda x : x[0:4])
top_studios['year_release'] = top_studios['year_release'].apply(lambda x : int(x))

top_studios = top_studios.sort_values(by='tomatometer_rating', ascending=False)
top_studios.rename(columns={'movie_title': 'Movie Title','tomatometer_rating': 'Rotten Tomatoes rating', 'studio_x':'Studio', 'year_release':'Year'}, inplace=True) 
top_studios
#print(type(top_studios['year_release']))
```

```python
top_studios.to_csv('kids_movies.csv',index = False)
```
